###2019

```python
file_path = "/content/drive/MyDrive/쿠다 1조 데이터셋/2019_최종_집값_정리본_물가O.csv"
output_file_path = "/content/drive/MyDrive/쿠다 1조 데이터셋/2019 2020 조인/오피스텔(전월세)_실거래가_2019_조인전.csv"

data = pd.read_csv(file_path, encoding='euc-kr')

# 도로명주소와 전용면적(㎡) 열이 같은 중복된 데이터 중 랜덤으로 하나만 남기기
data_unique = data.drop_duplicates(subset=['도로명', '전용면적(㎡)'], keep='first')

# '시군구' 셀의 두 번째 단어까지와 '도로명' 셀을 합쳐서 '도로명 주소' 열 생성
data_unique['도로명 주소'] = data_unique['시군구'].str.split().str[0] + ' ' + data_unique['시군구'].str.split().str[1] + ' ' + data_unique['도로명']

# 새로운 파일로 저장
data_unique.to_csv(output_file_path, index=False, encoding='euc-kr')

print("새로운 파일이 저장되었습니다:", output_file_path)
```

```python
# 파일 경로 지정
file_path1 = "/content/drive/MyDrive/쿠다 1조 데이터셋/2019 2020 조인/오피스텔(전월세)_실거래가_2019_조인전.csv"
file_path2 = "/content/drive/MyDrive/쿠다 1조 데이터셋/위도 경도 변환 데이터/서울특별시 오피스텔 지오코딩(2023).csv"
output_file_path = "/content/drive/MyDrive/쿠다 1조 데이터셋/2019 2020 조인/최종2019집값.csv"

# CSV 파일 불러오기
data1 = pd.read_csv(file_path1, encoding='euc-kr')
data2 = pd.read_csv(file_path2, encoding='utf-8')

# 두 데이터프레임을 조인하여 새로운 데이터프레임 생성
merged_data = pd.merge(data1, data2, left_on='도로명 주소', right_on='표준신주소', how='inner')

# 새로운 파일로 저장
merged_data.to_csv(output_file_path, index=False, encoding='euc-kr')

print("조인된 파일이 저장되었습니다:", output_file_path)
```

###2019 ~ 2023 데이터 시계열 정리

```python
import pandas as pd

# 파일 경로 지정
file_path_2023 = "/content/drive/MyDrive/쿠다 1조 데이터셋/2023_최종_집값_정리본.csv"
output_file_path = "/content/drive/MyDrive/쿠다 1조 데이터셋/최종_집값_시계열정리(수정).csv"

# CSV 파일 불러오기
df_2023 = pd.read_csv(file_path_2023, encoding='euc-kr')

# 연도별 데이터를 리스트에 저장
df_list = []
for year in range(2019, 2023):
    file_path_year = f"/content/drive/MyDrive/쿠다 1조 데이터셋/{year}_최종_집값_정리본_물가O.csv"
    df_year = pd.read_csv(file_path_year, encoding='euc-kr')

    # 각 연도의 열에 접두사 추가
    df_year = df_year.rename(columns={'층': f'층_{year}', '최종 집값(만원)': f'최종 집값(만원)_{year}'})

    # 리스트에 추가
    df_list.append(df_year)

# 2023 데이터에 각 연도의 데이터를 조인
result_df = df_2023
for year, df_year in zip(range(2019, 2023), df_list):
    result_df = pd.merge(result_df, df_year[['표준신주소', '전용면적(㎡)', f'층_{year}', f'최종 집값(만원)_{year}']],
                         on=['표준신주소', '전용면적(㎡)'], how='left', suffixes=('_2023', f'_{year}'))

# 결과를 CSV 파일로 저장
result_df.to_csv(output_file_path, index=False, encoding='euc-kr')

print("병합된 파일이 저장되었습니다:", output_file_path)
```

```python
# 주소지와 면적이 같다면 중복 제거
result_unique = result_df.drop_duplicates(subset=['도로명 주소', '전용면적(㎡)'], keep='first')

# 2019년부터 2022년까지의 '최종 집값' 또는 '층' 열에 Null 값이 있는 행 제거
columns_to_check_null = [f'최종 집값(만원)_2019', f'최종 집값(만원)_2020', f'최종 집값(만원)_2021', f'최종 집값(만원)_2022', '층_2019', '층_2020', '층_2021', '층_2022']
result2 = result_unique.dropna(subset=columns_to_check_null)

# 파일로 저장
output_file_path = "/content/drive/MyDrive/쿠다 1조 데이터셋/집값_시계열변화_공백및중복제거(수정수정).csv"
result2.to_csv(output_file_path, index=False, encoding='euc-kr')

print("중복이 제거된 파일이 저장되었습니다:", output_file_path)
```
