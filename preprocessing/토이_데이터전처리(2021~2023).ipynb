```python
# 쉼표(,)를 제거하고 보증금(만원)과 월세금(만원)을 숫자로 변환하여 계산
def calculate_final_value(row):
    deposit = row['보증금(만원)']
    rent = row['월세금(만원)']
    if row['전월세구분'] == '전세':
        return float(deposit.replace(',', ''))
    else:
        rent_value = float(rent) if isinstance(rent, int) or (isinstance(rent, float) and not pd.isna(rent)) else 0
        deposit_value = float(deposit.replace(',', '')) if isinstance(deposit, int) or (isinstance(deposit, float) and not pd.isna(deposit)) else 0
        return (rent_value * 12 * 100 / 6) + deposit_value

data3['최종 집값'] = data3.apply(calculate_final_value, axis=1)
```

## 3. 연도별 최종집값 데이터 조인하여 시계열 변화 데이터 만들기

```python
# 각 연도별로 '최종 집값' 열 추출하여 데이터프레임 생성
dfs = []
for file_path in file_paths:
    try:
        year = file_path.split('/')[-1].split('_')[0]
        df = pd.read_csv(file_path, encoding='euc-kr')
        df = df.rename(columns={'최종 집값': f'최종 집값_{year}'})  # 연도를 열 이름 뒤에 추가
        dfs.append(df)
    except UnicodeDecodeError:
        print(f"UnicodeDecodeError: 파일을 읽을 수 없습니다: {file_path}")

# 각 연도별 데이터프레임을 조인하여 한 개의 데이터프레임으로 병합
result = dfs[0]  # 첫 번째 데이터프레임을 기준으로 설정
for df in dfs[1:]:
    result = pd.merge(result, df, on=['표준신주소', '전용면적(㎡)'], how='inner')

# 파일로 저장
output_file_path = "/Users/jeongnanju/Desktop/조인만하자/집값_시계열변화.csv"
result.to_csv(output_file_path, index=False, encoding='euc-kr')

print("파일이 저장되었습니다:", output_file_path)
```

```python
# 주소지와 면적이 같다면 중복 제거
result_unique = result.drop_duplicates(subset=['도로명 주소', '전용면적(㎡)'], keep='first')

# 파일로 저장
output_file_path = "/Users/jeongnanju/Desktop/조인만하자/집값_시계열변화_중복제거.csv"
result_unique.to_csv(output_file_path, index=False, encoding='euc-kr')

print("중복이 제거된 파일이 저장되었습니다:", output_file_path)
```
