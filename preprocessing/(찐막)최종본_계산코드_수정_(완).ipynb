```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')

data2023 = '/content/drive/MyDrive/쿠다 1조 데이터셋/오피스텔 실거래가(서울 전체)/연도별 실거래가 정리_난주/최종2023집값_정리본.csv'
data2022 = '/content/drive/MyDrive/쿠다 1조 데이터셋/오피스텔 실거래가(서울 전체)/연도별 실거래가 정리_난주/최종2022집값_정리본.csv'
data2021 = '/content/drive/MyDrive/쿠다 1조 데이터셋/오피스텔 실거래가(서울 전체)/연도별 실거래가 정리_난주/최종2021집값_정리본.csv'
data2020 = '/content/drive/MyDrive/쿠다 1조 데이터셋/오피스텔 실거래가(서울 전체)/2020/오피스텔(전월세)_실거래가_(2020).csv'
data2019 = '/content/drive/MyDrive/쿠다 1조 데이터셋/오피스텔 실거래가(서울 전체)/2019/오피스텔(전월세)_실거래가_(2019).csv'

df2023 = pd.read_csv(data2023,  encoding='cp949')
df2022 = pd.read_csv(data2022,  encoding='cp949')
df2021 = pd.read_csv(data2021,  encoding='cp949')
df2020 = pd.read_csv(data2020,  encoding='cp949')
df2019 = pd.read_csv(data2019,  encoding='cp949')

print(df2023.head())
print(df2022.head())
print(df2021.head())
print(df2020.head())
print(df2019.head())
```

```python
def calculate_final_value2023(row):
    deposit = row['보증금(만원)']
    rent = row['월세금(만원)']
    if row['전월세구분'] == '전세':
        return float(deposit.replace(',', ''))
    else:
        deposit_value = float(deposit.replace(',', '')) if str(deposit).replace(',', '').isnumeric() else 0
        rent_value = float(rent) if str(rent).isnumeric() else 0
        return round(((rent_value * 12) + (0.054 * deposit_value)) / 0.054, -2)

def calculate_final_value2022(row):
    deposit = row['보증금(만원)']
    rent = row['월세금(만원)']
    if row['전월세구분'] == '전세':
        return  round(float(deposit.replace(',', ''))*(1.014), -2)
    else:
        deposit_value = float(deposit.replace(',', '')) if str(deposit).replace(',', '').isnumeric() else 0
        rent_value = float(rent) if str(rent).isnumeric() else 0
        fin_rent_value=(((rent_value * 12) + (0.048 * deposit_value)) / 0.048)
        fin_rent_value2022=(fin_rent_value)*1.014
        return round(fin_rent_value2022, -2)

def calculate_final_value2021(row):
    deposit = row['보증금(만원)']
    rent = row['월세금(만원)']
    if row['전월세구분'] == '전세':
        return round(float(deposit.replace(',', ''))*1.040364, -2)
    else:
        deposit_value = float(deposit.replace(',', '')) if str(deposit).replace(',', '').isnumeric() else 0
        rent_value = float(rent) if str(rent).isnumeric() else 0
        fin_rent_value=(((rent_value * 12) + (0.047 * deposit_value)) / 0.047)
        fin_rent_value2021=(fin_rent_value)*1.040364  #1.026*1.014=1.040364
        return round(fin_rent_value2021, -2)

def calculate_final_value2020(row):
    deposit = row['보증금(만원)']
    rent = row['월세금(만원)']
    if row['전월세구분'] == '전세':
        return round(float(deposit.replace(',', ''))*(1.08501), -2)
    else:
        deposit_value = float(deposit.replace(',', '')) if str(deposit).replace(',', '').isnumeric() else 0
        rent_value = float(rent) if str(rent).isnumeric() else 0
        fin_rent_value=(((rent_value * 12) + (0.05 * deposit_value)) / 0.05)
        fin_rent_value2020=(fin_rent_value)*(1.08501)  #1.026*1.014*1.043=1.08501
        return round(fin_rent_value2020, -2)

def calculate_final_value2019(row):
    deposit = row['보증금(만원)']
    rent = row['월세금(만원)']
    if row['전월세구분'] == '전세':
        return round(float(deposit.replace(',', ''))*(1.0775), -2)
    else:
        deposit_value = float(deposit.replace(',', '')) if str(deposit).replace(',', '').isnumeric() else 0
        rent_value = float(rent) if str(rent).isnumeric() else 0
        fin_rent_value=(((rent_value * 12) + (0.052 * deposit_value)) / 0.052)
        fin_rent_value2019=(fin_rent_value)*(1.0775)  #1.026*1.014*1.043*0.993=1.0775
        return round(fin_rent_value2019, -2)

# 최종 집값 계산 및 새로운 열 추가
df2023['최종 집값(만원)'] = df2023.apply(calculate_final_value2023, axis=1)
df2022['최종 집값(만원)'] = df2022.apply(calculate_final_value2022, axis=1)
df2021['최종 집값(만원)'] = df2021.apply(calculate_final_value2021, axis=1)
df2020['최종 집값(만원)'] = df2020.apply(calculate_final_value2020, axis=1)
df2019['최종 집값(만원)'] = df2019.apply(calculate_final_value2019, axis=1)

# 결과 확인
print(df2023.head())
```
